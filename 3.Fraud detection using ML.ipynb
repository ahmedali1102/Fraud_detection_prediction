```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import  matplotlib.pyplot as plt 
%matplotlib inline
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression 

sns.set_style('darkgrid')
```

```python
## Dataset 
df = pd.read_csv('payment_fraud.csv')
```

```python
df.head()
```

```python
df.isnull().sum()## checking the null valeus 
```

```python
paymthd = df.paymentMethod.value_counts()
plt.figure(figsize=(5, 5))
sns.barplot(x = paymthd.index,y = paymthd);
plt.ylabel('Count');
```

```python
df.paymentMethod.value_counts()
```

```python
df.label.value_counts() ## count the number of 0's and 1's
```

```python
## coverting paymentMethod column into label encoding
paymthd_label = {v:k for k, v in enumerate(df.paymentMethod.unique())}
df.paymentMethod = df.paymentMethod.map(paymthd_label)
```

```python
df.head()
```

```python
## corr(): it gives the correlation between the featuers
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(), annot= True);
```

```python
df.describe()
```

```python
## independent and dependent features
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
## scaling 

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
## train test split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
print("y_train shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

```python
## logisticRegression Model
lg = LogisticRegression()

## training
lg.fit(X_train, y_train)
```

```python
## prediction 
pred = lg.predict(X_test)
```

```python
print("----------------------------------------------------Accuracy------------------------------------------------------")
print(accuracy_score(y_test, pred))
print()

print("---------------------------------------------------Classification Report---------------------------------------------")
print(classification_report(y_test, pred))
print()

print("-------------------------------------------------Confustion Metrics----------------------------------------------------")
plt.figure(figsize=(5, 5));
sns.heatmap(confusion_matrix(y_test, pred), annot=True, fmt='g');
```

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
os = RandomOverSampler()
X_train_os, y_train_os = os.fit_resample(X_train, y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_os)))
```

```python
## logisticRegression Model
lg_os = LogisticRegression()

## training
lg.fit(X_train_os, y_train_os)
```

```python
## prediction 
pred = lg.predict(X_train_os)
```

```python
print("----------------------------------------------------Accuracy------------------------------------------------------")
print(accuracy_score(y_train_os, pred))
print()

print("---------------------------------------------------Classification Report---------------------------------------------")
print(classification_report(y_train_os, pred))
print()

print("-------------------------------------------------Confustion Metrics----------------------------------------------------")
plt.figure(figsize=(5, 5));
sns.heatmap(confusion_matrix(y_train_os, pred), annot=True, fmt='g');
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train_os, y_train_os)
```

```python
clf_pred = clf.predict(X_train_os)
```

```python
print("----------------------------------------------------Accuracy------------------------------------------------------")
print(accuracy_score(y_train_os, clf_pred))
print()

print("---------------------------------------------------Classification Report---------------------------------------------")
print(classification_report(y_train_os, clf_pred))
print()

print("-------------------------------------------------Confustion Metrics----------------------------------------------------")
plt.figure(figsize=(5, 5));
sns.heatmap(confusion_matrix(y_train_os, clf_pred), annot=True, fmt='g');
```
